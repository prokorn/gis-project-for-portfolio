```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.tree import plot_tree
```

```python
file1 = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df = file1
```

```python
x = df.drop('species', axis=1)
y = df['species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
score = accuracy_score(y_test, predictions)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```

```python
mystery_flower = [[1.1, 6.5, 2.4, 0.2]]
myst = model.predict(mystery_flower)
print(myst)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15)) # This makes the drawing bigger!
plot_tree(model,
          feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
          class_names=['setosa', 'versicolor', 'virginica'],
          filled=True)
plt.show()
```
